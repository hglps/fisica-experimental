# Plot dos gr√°ficos dos resistores

```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
data_47 = pd.read_excel('47ohm.xlsx')
data_100 = pd.read_excel('100ohm.xlsx')
data_200 = pd.read_excel('200ohm.xlsx')
data_2k = pd.read_excel('2Kohm.xlsx')
data_560k = pd.read_excel('560Kohm.xlsx')
```

```python
fig = plt.figure(figsize=(10,5))
plt.tight_layout

x1 = data_47['Corrente']
y1 = data_47['Tens√£o']

ax1 = fig.add_axes([0.1, 0.1, 0.1, 0.25])
ax2 = fig.add_axes([0.4, 0.1, 0.1, 0.25])
ax3 = fig.add_axes([0.7, 0.1, 0.1, 0.25])
ax4 = fig.add_axes([0.1, 0.6, 0.1, 0.25])
ax5 = fig.add_axes([0.5, 0.6, 0.1, 0.25])

ax1.plot(x1,y1)
ax1.set_xlabel('Corrente (mA)')
ax1.set_ylabel('Tens√£o (V)')
ax1.set_title('Gr√°fico 8 - Resistor de 47 Ohm\nFundo de escala 200mA')

x2 = data_100['Corrente']
y2 = data_100['Tens√£o']

ax2.plot(x2,y2,color='r')
ax2.set_xlabel('Corrente (mA)')
ax2.set_ylabel('Tens√£o (V)')
ax2.set_title('Gr√°fico 9 - Resistor de 100 Ohm\nFundo de escala 200mA')


x3 = data_200['Corrente']
y3 = data_200['Tens√£o']

ax3.plot(x3,y3,color='g')
ax3.set_xlabel('Corrente (mA)')
ax3.set_ylabel('Tens√£o (V)')
ax3.set_title('Gr√°fico 10 - Resistor de 200 Ohm\nFundo de escala 200mA')

x4 = data_2k['Corrente']
y4 = data_2k['Tens√£o']

ax4.plot(x4,y4,color='y')
ax4.set_xlabel('Corrente (mA)')
ax4.set_ylabel('Tens√£o (V)')
ax4.set_title('Gr√°fico 6 - Resistor de 2 KOhm\nFundo de escala 20mA')

x5 = data_560k['Corrente']
y5 = data_560k['Tens√£o']

ax5.plot(x5,y5,color='b')
ax5.set_xlabel('Corrente (mA)')
ax5.set_ylabel('Tens√£o (V)')
ax5.set_title('Gr√°fico 7 - Resistor de 560 KOhm\nFundo de escala 200uA')

plt.savefig('resistores.png')
```
