# Plots de carga e descarga de capacitor

```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
carga = pd.read_excel('carga.xlsx')
descarga = pd.read_excel('descarga.xlsx')
```

```python
carga
```

```python
descarga
```

```python
fig = plt.figure(figsize=(10,10))

carga_t = carga['Tempo']
carga_v = carga['Tensão']

descarga_t = descarga['Tempo']
descarga_v = descarga['Tensão']

ax1 = fig.add_axes([0.1, 0.1, 0.9, 0.8])
ax1.plot(carga_t,carga_v)
ax1.set_title('Gráfico 6 - Carga e Descarga de Capacitor')
ax1.set_xlabel("Tempo (s)")
ax1.set_ylabel("Tensão (V)")

ax2 = fig.add_axes([0.1, 0.1, 0.9, 0.8])
ax2.plot(descarga_t,descarga_v,color='r')
ax1.legend(['Carga','Descarga'])

plt.savefig('capacitor.png')
```
